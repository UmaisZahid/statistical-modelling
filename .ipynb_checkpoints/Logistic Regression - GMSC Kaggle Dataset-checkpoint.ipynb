# Model Description

Possibly the simplest, non-trivial, model you can apply to a classification task: logistic regression is a generalisation of ordinary least squares (OLS) models to response variables having a binomial distribution. It is member of the family of generalised linear models, which generalise OLS to response variables following any distribution in the exponential family. 

In this case our response variable will be the probability of default (serious delinquency) within 2 years. A serious delinquency is any time when a loan or line of credit is more than 90 days past due. 

Our model will assume that the log odds of this event is equal to a linear combination of input features. I.e: 
$$ \log(\frac{p}{1-p}) = \beta X$$

Where $\beta$ and $X$ are a vector of coefficients and input features respectively. 

Rearranging this tells us that this is equivalent to saying the probability of the response being true is equal to the logistic/sigmoid function applied to the linear combination of our input features.

```python
#Import statements
%matplotlib inline
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
pd.options.display.max_rows = 20
from sklearn.model_selection import train_test_split
from sklearn import metrics
from statsmodels.stats.outliers_influence import variance_inflation_factor

#Import raw data
data_raw = pd.read_csv("./Datasets/Kaggle - Give Me Some Credit/cs-training.csv")
output_raw = pd.read_csv("./Datasets/Kaggle - Give Me Some Credit/cs-test.csv")
data_clean = data_raw.copy()
output_clean = output_raw.copy()
# Set seed so train/test split remains consistent over runs
seed = 42
```

## Data Cleaning

*Good hygiene  is essential when handling raw data.*

```python
data_raw.head()
```

We impute in our NaNs with the datasets means. We can also later try removing them entirely to see if the BIC/AIC of the model improves (decreases). 

```python
# Find out how many NaNs we're dealing with here. 
print(data_raw.isnull().sum())

# Fill NaNs with mean.
data_clean['MonthlyIncome'].fillna(data_clean['MonthlyIncome'].mean(),inplace=True)
data_clean['NumberOfDependents'].fillna(data_clean['NumberOfDependents'].mean(),inplace=True)
#Number of delinquents vs non
print((data_clean['SeriousDlqin2yrs'] == 0).value_counts())

# Split data to train and test
train, test = train_test_split(data_clean,test_size=0.3,random_state=seed)
X = train.iloc[:,2:]
Y = train.iloc[:,1]
X_test = test.iloc[:,2:]
Y_test = test.iloc[:,1]
```

```python
f, axes = plt.subplots(2, 5, figsize=(17, 8))
data_describe = data_clean.describe()
for index,ax in enumerate(axes.reshape(-1)):
    delinquent = data_clean['SeriousDlqin2yrs'] == 1
    sns.distplot(data_clean.loc[delinquent].iloc[:,index+2],ax=ax,color="red",label="Delinquent")
    sns.distplot(data_clean.loc[~delinquent].iloc[:,index+2],ax=ax,color="blue",label="Not Delinquent")
    ax.legend()
```

```python
# Get skews
print("Skew values: ")
skews = data_clean.iloc[:,2:].skew(axis=0)
print(skews)
```

## Check collinearity

```python
# Check collinearity
corr = data_clean.corr()
fig, ax = plt.subplots(figsize=(10,8))
sns.heatmap(corr,annot=True,ax=ax)
```

### Note: 

As expected, the features relating to deliquencies in various time bins are highly correlated with each other. We also have correlation between the two open credit line variables, also expected since NumberOfOpenCreditLinesAndLoans is inclusive of all loans, including mortgage or real estate loans. 

Collinearity tends to cause overinflation of the variance (standard error) of our regression coefficients. To check that collinearity is an actual problem we will need to calculate the VIF (variance inflation factor) which quantifies just how large this standard error is. A VIF of > 10 is a sign that multicollinearity is high. We will check on this later once we develop our models. Unfortunately VIF for logistic regression isn't particularly well-defined as it is designed for OLS with the R^2 metric. Hypothetically you could use Pseudo-R^2 but as I am using this simply as a diagnostic tool I will use standard OLS based VIF. 

```python
def getVIFs(design_matrix):
    output = pd.DataFrame(0,columns=['VIFs'],index=design_matrix.columns)
    for index,column in enumerate(design_matrix.columns): 
        output['VIFs'][column] = variance_inflation_factor(design_matrix.to_numpy(),index)
    return output


getVIFs(X_test)
```

# Logistic Regression Model - Initial Attempt

We use all the available features for our input feature vector and train a regression model on the P(SeriousDlqin2Yrs)


```python
model1 = LogisticRegression(random_state=seed,solver='lbfgs',penalty='none',max_iter=2000).fit(X,Y)
```

```python
def print_metrics(realX,realY,model):
    predicted = model.predict(realX)
    predicted_prob = model.predict_proba(realX)[:,1]
    confusion_matrix = model1.score
    fpr, tpr, _ = metrics.roc_curve(realY,predicted_prob)
    auc = metrics.roc_auc_score(realY,predicted_prob)
    f1score = metrics.f1_score(realY,predicted)
    pscore = metrics.precision_score(realY,predicted)
    rscore = metrics.recall_score(realY,predicted)
    print("The AUC is: ", auc)
    print("The F1 score is: ",f1score)
    print("The precision score is: ", pscore)
    print("The recall score is: ", rscore)
    plt.plot(fpr,tpr)
    
```

```python
print_metrics(X_test,Y_test,model1)
```

## Results

Our first logistic regression model has given us pretty poor performance. It's better than chance, but not significantly so. Our AUC is only 0.691, with an ideal value being 1, similarly our F1 score is ~0.0782, with an ideal value also taking 1. 

Possible improvements: 
- Deal with right skew of our dependent variables
- Deal with multicollinearity between our delinquency dependent variables (NumberOfTimes...)

Let's implement these in turn and see what improvements we get returned. 

# Logistic Regression Model - Attempt 2

We will deal with our skew by applying log transforms to all those dependents who have skew larger than 2.

```python
X_2 = X.copy()
X_test_2 = X_test.copy()

X_2.loc[:,skews.values>2] = np.log(X_2.loc[:,skews.values>2] + 0.00001)
X_test_2.loc[:,skews.values>2] = np.log(X_test_2.loc[:,skews.values>2] + 0.00001)
```

```python
model2 = LogisticRegression(random_state=seed,solver='lbfgs',penalty='none',max_iter=2000).fit(X_2,Y)
```

```python
print_metrics(X_test_2,Y_test,model2)
```

## Results

Nice! Major improvement on our ROC curve. Our new AUC is 0.837, and our F1 score has more than tripled!

Now let's try combining our collinear features into 1 and seeing if we get an improvement

# Logistic Regression Model - Attempt 3

In this model we will follow model 2's approach but first combine all 3 highly collinear delinquency dependents into one variable. 

```python
# Combine columns
X_3 = X.copy()
X_test_3 = X_test.copy()
X_3
combined_cols = ['NumberOfTime30-59DaysPastDueNotWorse','NumberOfTime60-89DaysPastDueNotWorse','NumberOfTimes90DaysLate']

X_3['CombinedDeliquency'] = X_3[combined_cols[0]] + X_3[combined_cols[1]] + X_3[combined_cols[2]]
X_test_3['CombinedDeliquency'] = X_test_3[combined_cols[0]] + X_test_3[combined_cols[1]] + X_test_3[combined_cols[2]]
X_3 = X_3.drop(combined_cols,axis=1)
X_test_3 = X_test_3.drop(combined_cols,axis=1)


# Calculate new skews including combined columns
new_skews = X_3.skew(axis=0)

# Apply log to columns with skew values greater than 2
X_3.loc[:,new_skews.values>2] = np.log(X_3.loc[:,new_skews.values>2] + 0.00001)
X_test_3.loc[:,new_skews.values>2] = np.log(X_test_3.loc[:,new_skews.values>2] + 0.00001)
```

```python
model3 = LogisticRegression(random_state=seed,solver='lbfgs',penalty='none',max_iter=2000).fit(X_3,Y)
```

```python
print_metrics(X_test_3,Y_test,model3)
```

## Results

Yikes! Looks like a decrease in all metrics. Lower AUC, F1, precision AND recall. Looks like the collinearity induced coefficient variance is a worthy sacrifice for a better model.
